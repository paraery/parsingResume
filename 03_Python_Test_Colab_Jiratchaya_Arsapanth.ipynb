```python
def readLineText(fileName):
     y=[]
     with open(fileName, 'r') as f:
            for line in f:
                 y.append(line.split())
     return y
```

```python
import re
def clean_txt(txt):
    txt = re.sub(r'[^\w\s:()@.]|:(?![()])|(?<!:)[()]', '', txt)
    txt = txt.lower()
    txt = re.sub(r'\n',' ', txt)
    txt = re.sub(r'  ',' ',txt)
    return txt
```

```python
import pandas as pd
```

```python
textlist = readLineText('input.txt')
for txt in textlist:
  clean_txt(txt)
df = pd.DataFrame(textlist)
df.to_excel('clean_output.xlsx')
```
